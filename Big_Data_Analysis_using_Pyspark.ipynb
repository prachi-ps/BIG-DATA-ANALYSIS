## **10. Create new column**
Creating column 'Usage_Level' which tells how much the car had driven based on 'Kms_Driven'. Tells whether the car is Lightly used, Moderately used, or Heavily used.

```python
from pyspark.sql.functions import col, when

df = df.withColumn(
    "Usage_Level",
    when(col("Kms_driven") <= 10000, "Lightly used")
    .when(col("Kms_Driven") <= 50000, "Moderately used")
    .otherwise("Heavily used")
)
df.show()
```

## **11. Rename 'Year' column and give new name 'Manufacture_Year'**

```python
df = df.withColumnRenamed('Year', 'Manufacture_Year')
df.show()
```

## **12. Display name of cars having selling price greater than 5.50**

```python
df.filter(df['Selling_Price_Lakh'] > 5.50).select('Car_Name').show()
```

## **13. Display name and transmission of cars having selling price greater than 5.50**

```python
df.filter(df['Selling_Price_Lakh'] > 5.50).select(['Car_Name', 'Transmission']).show()
```

## **14. Display names of manual cars having selling price greater than 5.50**

```python
df.filter((df['Selling_Price_Lakh'] > 5.50) & (df['Transmission'] == 'Manual')).select('Car_Name').show()
```

## **15. Display names of automatic cars having selling price greater than 5.50**

```python
df.filter((df['Selling_Price_Lakh'] > 5.50) & ~(df['Transmission'] == 'Manual')).select('Car_Name').show()
```

## **16. Display average selling price of manual and automatic cars**

```python
from pyspark.sql.functions import avg

df.groupby('Transmission').agg(avg('Selling_Price_Lakh')).show()
```

## **17. Display all cars which run on petrol fuel and were bought from a dealer**

```python
df.filter((df['Fuel_Type'] == 'Petrol') & (df['Seller_Type'] == 'Dealer')).select('Car_Name').show()
```

## **18. Display the maximum and minimum present price of cars, grouped by fuel type**

```python
from pyspark.sql.functions import min, max

df.groupby('Fuel_Type').agg(min('Present_Price_Lakh'), (max('Present_Price_Lakh'))).show()
```

## **19. Display all cars sorted by selling price in descending order**

```python
df.orderBy(df['Selling_Price_Lakh'].desc()).show()
```

## **20. Display total number of cars grouped by seller type**

```python
from pyspark.sql.functions import count

df.groupby('Seller_Type').agg(count('Car_Name')).show()
```
